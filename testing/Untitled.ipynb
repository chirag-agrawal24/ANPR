```python
from sklearn.metrics import confusion_matrix
```

```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
l=[[109,5],[14,0]]
df=pd.DataFrame(l,columns=['Number_plate','Background'],index=['Number_plate','Background'])
```

```python
import matplotlib.pyplot as plt
```

```python
sns.heatmap(df,annot=True,fmt='.3g',cmap='Blues')
```

```python
sns.heatmap(df/np.sum(df),annot=True,fmt='.3g',cmap='Blues')
```

```python
ax = sns.heatmap(df, annot=True, fmt='.3g', cmap='Blues')

# Setting the axis labels
ax.set_xlabel('Truth')
ax.set_ylabel('Predicted')
plt.savefig('confusion_matrix.png')
# Display the heatmap
plt.show()
```

```python
ax = sns.heatmap(df/np.sum(df.to_numpy(),axis=None), annot=True, fmt='.3g', cmap='Blues')

# Setting the axis labels
ax.set_xlabel('Truth')
ax.set_ylabel('Predicted')

# Display the heatmap
plt.savefig('cm_normalized_all.png')
plt.show()
```

```python
precision=df.iloc[0,0]/(df.iloc[0,0]+df.iloc[0,1])
recall=df.iloc[0,0]/(df.iloc[0,0]+df.iloc[1,0])
f1_score=2*precision*recall/(precision+recall)
```

```python
sns.barplot([precision,recall,f1_score])
```
